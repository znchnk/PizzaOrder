```python
#Анализ чувствительности (NPV,, PI, DPP, ROI, IRR ....)
#NPV
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits .mplot3d import Axes3D
```

```python
investment = [-1]*6
income = [0.25]*30
cash_flow = investment +income
```

#### Расчет NPV инвестиционного проекта

```python
NPV = np.npv(0.05, cash_flow)
NPV
```

```python
inc_mass = np.arange(0.05, 1, 0.01) 
npv_inc1 = [ round(np.npv(0.05, [-1]*6+[i]*30)  , 2) for i in inc_mass ]
```

```python
plt.plot(inc_mass, npv_inc1)
plt.xlabel("income")
plt.ylabel("NPV")
plt.title('Зависимость NPV от величины дохода при инветициях = 1 млн/месяц')

plt.show()
```

```python
inv_mass = np.arange(0, 10, 0.01) 
npv_inc2 = [ round(np.npv(0.05, [-i]*6+[0.25]*30)  , 2) for i in inv_mass ]
plt.plot(inv_mass, npv_inc2)
plt.xlabel("investment")
plt.ylabel("NPV")
plt.title('Зависимость NPV от величины инвестиций при доходе = 250 тыс/месяц')
plt.show()
```

```python
x = np.arange(0, 10, 0.2) #инвестиции
y = np.arange(0, 2, 0.05) #доход
xgrid , ygrid = np.meshgrid(x, y)
zgrid2 = []
for i in y:
    zgrid2.append([round(np.npv(0.05, [-ii]*6+[i]*30), 2) for ii in x])
zgrid = np.array(zgrid2)
#zgrid3 = 0*xgrid  + 0* ygrid 

frame = plt.figure(figsize =(10 ,8))
axes = Axes3D(frame)
axes.plot_surface (xgrid , ygrid , zgrid , cmap=plt.cm.Pastel2_r , linewidth =0.5 , rstride =1, cstride =1)
#axes.plot_surface (xgrid , ygrid , zgrid3 , color="lightyellow")
axes.set_xlabel ('investment')
axes.set_ylabel ('income')
axes.set_zlabel ('NPV')
plt.title('Зависимость NPV от величины инвестиций и дохода')
plt.show ()
```

#### Расчет IRR инвестиционного проекта

```python
#
IRR = np.irr(cash_flow)
IRR
```

```python
inc_mass = np.arange(0.005, 2, 0.005) 
irr_inc1 = [ round(np.irr( [-1]*6+[i]*30 )  , 2) for i in inc_mass ]

plt.plot(inc_mass, irr_inc1)
plt.xlabel("income")
plt.ylabel("IRR")
plt.title('Зависимость IRR от величины дохода при инветициях = 1 млн/месяц')
plt.show()
```

```python
irr_mass = np.arange(0, 10, 0.1) 
irr_inc2 = [ round(np.irr([-i]*6+[0.25]*30)  , 2) for i in irr_mass ]
plt.plot(irr_mass, irr_inc2)
plt.xlabel("investment")
plt.ylabel("IRR")
plt.title('Зависимость IRR от величины инвестиций при доходе = 250 тыс/месяц')
plt.show()
```

```python
x = np.arange(0.2, 10, 0.2) #инвестиции
y = np.arange(0.05, 2, 0.05) #доход
xgrid , ygrid = np.meshgrid(x, y)
zgrid2 = []
for i in y:
    zgrid2.append([round(np.irr([-ii]*6+[i]*30), 2) for ii in x])
zgrid = np.array(zgrid2)
#zgrid3 = 0*xgrid  + 0*ygrid 

frame = plt.figure(figsize =(10 ,8))
axes = Axes3D(frame)
axes.plot_surface (xgrid , ygrid , zgrid , cmap=plt.cm.Pastel2_r , linewidth =0.5 , rstride =1, cstride =1)
#axes.plot_surface (xgrid , ygrid , zgrid3 , color="lightyellow")
axes.set_xlabel ('investment')
axes.set_ylabel ('income')
axes.set_zlabel ('IRR')
plt.title('Зависимость IRR от величины инвестиций и дохода')
plt.show ()
```

### Расчет PI инвестиционного проекта

```python
PI = np.npv(0.05, cash_flow) / (np.array(investment)*-1).sum()
PI
```

```python
inc_mass = np.arange(0.05, 1.5, 0.01) 
#pi_inc1 = [ round(np.npv(0.05, [-1]*6+[i]*30) / (np.array(investment)*-1).sum() , 2) for i in inc_mass ]
pi_inc1 = [ round(np.npv(0.05, [-1]*6+[i]*30) / np.npv(0.05, [1]*6 ), 2) for i in inc_mass ]
```

```python
plt.plot(inc_mass, pi_inc1)
plt.xlabel("income")
plt.ylabel("PI")
plt.title('Зависимость PI от величины дохода при инветициях = 1 млн/месяц')
plt.show()
```

```python
inv_mass = np.arange(0.01, 10, 0.01) 
pi_inc2 = [ round(np.npv(0.05, [-i]*6+[0.25]*30) / np.npv(0.05, [i]*6 ), 2) for i in inv_mass ]
plt.plot(inv_mass, pi_inc2)
plt.xlabel("investment")
plt.ylabel("PI")
plt.title('Зависимость PI от величины инвестиций при доходе = 250 тыс/месяц')
plt.show()
```

```python
x = np.arange(0.2, 10, 0.2) #инвестиции
y = np.arange(0.05, 2, 0.05) #доход
xgrid , ygrid = np.meshgrid(x, y)
zgrid2 = []
for i in y:
    zgrid2.append([round(np.npv(0.05, [-ii]*6+[i]*30)  / np.npv(0.05, [ii]*6 ) , 2) for ii in x])
zgrid = np.array(zgrid2)
#zgrid3 = 0*xgrid  + 0* ygrid 

frame = plt.figure(figsize =(10 ,8))
axes = Axes3D(frame)
axes.plot_surface (xgrid , ygrid , zgrid , cmap=plt.cm.Pastel2_r , linewidth =0.5 , rstride =1, cstride =1)
#axes.plot_surface (xgrid , ygrid , zgrid3 , color="lightyellow")
axes.set_xlabel ('investment')
axes.set_ylabel ('income')
axes.set_zlabel ('PI')
plt.title('Зависимость PI от величины инвестиций и дохода')
plt.show ()
```
